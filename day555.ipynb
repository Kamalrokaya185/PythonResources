```python
import matplotlib.pyplot as plt

labels=['Group1','Group2','Group3','Group4','Group5']
men_age=[20,35,30,35,27]
women_age=[25,32,34,20,25]

fig,ax=plt.subplots()

ax.bar(labels,men_age,label='Men')
ax.bar(labels,women_age,bottom=men_age,label='Women')

ax.set_ylabel('Age')
ax.set_ylabel('Group Number')
ax.set_ylabel('Gender Ratio by group')

ax.legend()
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
classes=['python','R','AI','Ml',"Ds"]
classes1_student=[30,10,20,25,10]
classes2_student=[40,5,20,20,10]
class_index=np.arange(len(classes))
width=0.2

plt.bar(class_index,classes1_student,width,label='classes1 student')
plt.bar(class_index,classes2_student,width,label='classes2 student')

plt.xticks(class_index+width,classes)

plt.title('Bar charts of classes')
plt.xlabel("classes")
plt.ylabel("no.of student")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as  num
x_axis=num.array([0,5,10,15,20,25,30])
plt.plot(x_axis,x_axis**2,'red',label="x=y**2")

plt.legend()
plt.xlabel("x_Axis")
plt.ylabel("Y_Axis")

plt.title("My plot")
plt.show()
```

```python
y_axis = x_axis**2
empty_canvas = plt.figure()
create_axis = empty_canvas.add_axes([0,0,1,1]) 

create_axis.plot(x_axis, y_axis,'b')
create_axis = empty_canvas.add_axes([0,0,1,1]) 

create_axis.plot(x_axis, y_axis, 'b', label = "x = x**2")
create_axis.plot(x_axis, y_axis+(x_axis*3), 'g', label = "x = y+(x*3)")

create_axis.legend(loc=0)
```

```python
import matplotlib.pyplot as plt

data = {'Batch 1': 10, 'Batch 2': 15, 'Batch 3': 5, 'Batch 4': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1,3, figsize=(10, 3))

axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

fig.suptitle('Subplots')
```
